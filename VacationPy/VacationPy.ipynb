```python
def f_to_c(temp):
    return (temp - 32) * (5/9)
def mph_to_kmph(kmph):
    return kmph*1.609
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
```

```python
locations = weather_df[["Lat", "Lng"]].astype(float)
humidity = weather_df["Humidity"].astype(float)
```

```python
fig = gmaps.figure()

humidity_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)

fig.add_layer(humidity_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# A max temperature lower than 80 degrees but higher than 70.

# Wind speed less than 10 mph.

# Zero cloudiness.



weather_df.dtypes

narrow_weather_df = weather_df.loc[(weather_df["Max Temp"] < f_to_c(80)) 
                                   & (weather_df["Max Temp"] > f_to_c(70)) 
                                   & (weather_df["Wind Speed"] < mph_to_kmph(10)) 
                                   & (weather_df["Cloudiness"] == 0),:]
narrow_weather_df = narrow_weather_df.dropna()
narrow_weather_df

```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = narrow_weather_df
```

```python
hotel_df["Hotel Name"] = ""
```

```python
base_url_hotel = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

params_hotel = {
    "location": "",
    "radius" : 5000,
    "type": "lodging",
    "key": g_key
}

for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    
    params_hotel["location"] = f"{lat},{lng}"
    
    response_hotel = requests.get(base_url_hotel, params=params_hotel).json()
    
    try:
        hotel_df.loc[index, 'Hotel Name'] = response_hotel['results'][0]['name']
        
    except:
        print(f'Missing Hotel for city: {hotel_df.loc[index, "City"]}')

hotel_df

```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
marker_layer = gmaps.marker_layer(locations,info_box_content = hotel_info)
fig.add_layer(marker_layer)
fig

# Display Map
```
